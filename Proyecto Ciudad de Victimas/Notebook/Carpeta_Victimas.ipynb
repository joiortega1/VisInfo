## Gráfica comportamiento.
En la siguiente gráfica observaremos el comportamiento de la relación $registros/mes$ a lo largo del 2019

```python
def crecimiento_mes(data, mes, crecimiento):
    aux = 0
    for i in range(len(data)):
        if data.FechaInicio[i].split("/")[1] == mes:
            aux += 1
    crecimiento.append(aux)
```

```python
crecimiento_2019 = []
for i in range(5,13):
    if i <= 9:
        mes = '0'+str(i)
        crecimiento_mes(data, mes, crecimiento_2019)
    else: 
        mes = str(i)
        crecimiento_mes(data, mes, crecimiento_2019)
crecimiento_2019
```

```python
crecimiento_2020 = []
for i in range(1,13):
    if i <= 5 or i == 12:
        if i <= 5:
            mes = '0'+str(i)
            crecimiento_mes(data, mes, crecimiento_2020)
        else:
            mes = str(i)
            crecimiento_mes(data, mes, crecimiento_2020)
    else:
        pass
crecimiento_2020
```

```python
plt.imshow(corr)
plt.colorbar()
```

```python
import numpy as np
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set background color / chart style
sns.set_style(style = 'white')

# Set up  matplotlib figure
f, ax = plt.subplots(figsize=(10, 7))

# Add diverging colormap
cmap = sns.diverging_palette(220, 20, sep=20, as_cmap=True)

# Draw correlation plot
ax = sns.heatmap(corr, mask=mask, cmap=cmap, 
        square=True,
        linewidths=.5, cbar_kws={"shrink": .6}, ax=ax)
plt.savefig('correlacion.svg', transparent=True)
```
